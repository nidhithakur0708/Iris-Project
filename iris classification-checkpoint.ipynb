### Correlation Matrix

```python
#display the corelation matrix
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Now compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print(correlation_matrix)

```

```python
#plot the heat map
fig,ax=plt.subplots(figsize=(5,4))
sns.heatmap(correlation_matrix,annot=True,ax=ax,cmap='coolwarm')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pickle

#input data
X=df.drop(columns=['species'])
#output data
Y=df['species']
#split data in train and test
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30)
```

```python
#logistic regression
model=LogisticRegression()
model.fit(x_train,y_train)
print("Logistic Rgression Accuracy:",model.score(x_test,y_test)*100)
```

```python
#model training
model.fit(x_train.values,y_train.values)
```

```python
#print metric to get the performance
print("Accuracy: ",model.score(x_test,y_test)*100)
```

```python
#Knearest Neighbours
model=KNeighborsClassifier()
model.fit(x_train,y_train)
print("K-nearest Neighbours Accuracy:",model.score(x_test,y_test)*100)
```

```python
model.fit(x_train.values,y_train.values)
```

```python
#print metric to get the performance
print("Accuracy: ",model.score(x_test,y_test)*100)
```

```python
#decision tree
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
print("Decision Tree Accuracy Accuracy:",model.score(x_test,y_test)*100)
```

```python
model.fit(x_train.values,y_train.values)
```

```python
#print metric to get the performance
print("Accuracy: ",model.score(x_test,y_test)*100)
```

```python
#save the model
import pickle
filename='saved_model.sav'
try:
    with open(filename,'wb') as file:
        pickle.dump(model,file)
    print("Model saved Sucessfully")
except Exception as e:
    print(f"Error saving the model:{e}")
```
